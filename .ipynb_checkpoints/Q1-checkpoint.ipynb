```python
import numpy as np
from numpy import pi as PI
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sine(time_range, frequency, amplitude, phase):
    return amplitude * np.sin(2 * PI * frequency * time_range + phase)
```

# Sine/Complex Sine Waves

In this part, we'll plot some basic sine waves

```python
SAMPLE_RATE = 500
START_TIME = 0
END_TIME = 2
```

# Part 1A

```python
frequency = 3 #Hz
amplitude = 2
phase = PI/3
time_range = np.arange(START_TIME, END_TIME, 1/SAMPLE_RATE)

sinewave = amplitude * np.sin(2 * PI * frequency * time_range + phase)

plt.title("Sine Wave")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.plot(time_range, sinewave)
plt.show()
```

# Part 1B

```python
frequency = 3 #Hz
amplitude = 2
phase = PI/3
time_range = np.arange(START_TIME, END_TIME, 1/SAMPLE_RATE)


sinewave = amplitude * np.exp(2 * PI * 1.j * frequency * time_range + phase)

plt.title("Sine Wave")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.plot(time_range, np.imag(sinewave), '--r', time_range, np.real(sinewave), 'b')
plt.legend(["Imaginary", "Real"], loc="best")
plt.show()
```

Based on the above plot, we can conclude that the real part is the shifted plot of the imaginary part and vice-versa.

# Part 1C

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')

img = ax.plot(time_range, np.real(sinewave), np.imag(sinewave))
ax.set_title("3D Complex Sine Wave")
ax.set_xlabel("Time")
ax.set_ylabel("Real Part")
ax.set_zlabel("Imaginary Part")

plt.show()
```

The real and imaginary parts form a kind of spring shaped wave.

# Dot/Complex-Dot Product



# Part 2A 
[1]The statistical interpretation of dot product is the sum of the covariance of the vectors, this is a measure of how strongly two vectors share their pattern of values. <br>

If the vectors are identical every value is squared, if they differ the result will be smaller than the square of the larger value. This means that the resulting value is useful for measuring how similar two vectors are.<br>

If the two vectors are orthogonal they share no common variation and the sum of the products cancel out resulting in zero.

# Part 2B

```python
time_range = np.arange(START_TIME, END_TIME, 1/SAMPLE_RATE)
frequency = 5 #Hz
amplitude = 2
sinewave_1 = sine(time_range, frequency, amplitude, PI/2)
sinewave_2 = sine(time_range, frequency, amplitude, PI/2)

print("Dot product is:" ,np.dot(sinewave_1, sinewave_2))
```

```python
sinewave_2 = sine(time_range, frequency + 0.5, amplitude, PI/2)

print("Dot product is:" ,np.dot(sinewave_1, sinewave_2))
```

```python
sinewave_2 = sine(time_range, frequency - 0.5, amplitude, PI/2)

print("Dot product is:" ,np.dot(sinewave_1, sinewave_2))
```

```python
sinewave_2 = sine(time_range, frequency + 0.35, amplitude, PI/2)

print("Dot product is:" ,np.dot(sinewave_1, sinewave_2))
```

```python
sinewave_2 = sine(time_range, frequency - 0.35, amplitude, PI/2)

print("Dot product is:" ,np.dot(sinewave_1, sinewave_2))
```

By changing frequency, the wave shifts thus making the waves different and eventually decreasing the dot product result.

```python
sinewave_2 = sine(time_range, frequency, amplitude, PI/2 + PI/2)

print("Dot product is:" ,np.dot(sinewave_1, sinewave_2))
```

```python
sinewave_2 = sine(time_range, frequency, amplitude, PI/2 + PI/3)

print("Dot product is:" ,np.dot(sinewave_1, sinewave_2))
```

By increasing the phase, dot product decreases. This is obvious since the shift in the sine wave will decrease the dot product because the waves become less identical.

# Part 2C

In this part, we'll analyze a more complicated sine signal.<br>
The signal is 
$$sin(2\pi  5t+\theta)e^{-t^2}$$
Initially our sample rate is 1000 and $\theta$ is 0

```python
SAMPLE_RATE = 1000
START_TIME = -1
END_TIME = 1
```

```python
frequency = 5
amplitude = 1
theta = 0

time_range = np.arange(START_TIME, END_TIME, 1/SAMPLE_RATE)
f_t = amplitude * np.sin(2 * PI * frequency * time_range + theta) * np.exp(-(time_range**2))

plt.title("f(t)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(time_range, f_t)
plt.show()
```

```python
frequency_range = np.arange(2, 10.5, 0.5) # Changed to 10.5 to see max dot product in the next part
time_range = np.arange(START_TIME, END_TIME, 1/SAMPLE_RATE)
sinewaves = [np.dot(f_t, amplitude * np.sin(2 * PI * frequency * time_range + theta) * np.exp(-(time_range**2))) 
             for frequency in frequency_range]

plt.xlabel("Frequency")
plt.ylabel("Dot product")
plt.title("Dot product of f(t) and sinewaves")
plt.stem(frequency_range, sinewaves, use_line_collection=True)
plt.show()
```

The maximum product is for the wave with frequency 5 which is the identical frequency to our initial function.

```python
theta = PI/2
f_t = amplitude * np.sin(2 * PI * frequency * time_range + theta) * np.exp(-(time_range**2))

sinewaves = [np.dot(f_t, amplitude * np.sin(2 * PI * frequency * time_range + 0) * np.exp(-(time_range**2))) 
             for frequency in frequency_range]

plt.xlabel("Frequency")
plt.ylabel("Dot product")
plt.title("Dot product of f(t) with phase $\\frac{\pi}{2}$ and sinewaves")
plt.stem(frequency_range, sinewaves, use_line_collection=True)
plt.show()
```

```python
theta = PI/4
f_t = amplitude * np.sin(2 * PI * frequency * time_range + theta) * np.exp(-(time_range**2))

sinewaves = [np.dot(f_t, amplitude * np.sin(2 * PI * frequency * time_range + 0) * np.exp(-(time_range**2))) 
             for frequency in frequency_range]

plt.xlabel("Frequency")
plt.ylabel("Dot product")
plt.title("Dot product of f(t) with phase $\\frac{\pi}{4}$ and sinewaves")
plt.stem(frequency_range, sinewaves, use_line_collection=True)
plt.show()
```

As expected, by increasing phase, the dot product decreases. For $\frac{\pi}{4}$ shift, the maximum is still at frequency 5 but the value of maximum product is lower than phase 0. For $\frac{\pi}{2}$ shift, the maximum is still at frequency 5 but the value range for dot products are drastically different. The maximum value is 4 in this phase.<br>

In this part, we'll analyze the previous sine signal with complex sine waves.


```python
def plot_stems(f_t):
    sinewaves = [np.vdot(f_t, np.exp(2 * np.pi * 1.j * frequency * time_range)) for frequency in frequency_range]

    fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(10,3))
    axes[0].stem(frequency_range, np.real(sinewaves), label = "Real ($\\theta=0$)")
    axes[0].legend(loc="best")
    axes[0].set_xlabel("Frequency")
    axes[0].set_ylabel("Dot product")

    axes[1].stem(frequency_range, np.imag(sinewaves), label = "Imaginary ($\\theta=0$)")
    axes[1].legend(loc="best")
    axes[1].set_xlabel("Frequency")
    axes[1].set_ylabel("Dot product")

    plt.tight_layout()
    plt.show()
```

```python
theta = 0
f_t = amplitude * np.sin(2 * PI * frequency * time_range + theta) * np.exp(-(time_range**2))
plot_stems(f_t)
```

```python
theta = PI/2
f_t = amplitude * np.sin(2 * PI * frequency * time_range + theta) * np.exp(-(time_range**2))
plot_stems(f_t)
```

```python
theta = PI/4
f_t = amplitude * np.sin(2 * PI * frequency * time_range + theta) * np.exp(-(time_range**2))
plot_stems(f_t)
```

```python
frequency = 5
amplitude = 1
time_range = np.arange(START_TIME, END_TIME, 1/50)
theta_range = np.linspace(0, 7*PI/2, 100)

sinewaves = []
_amplitude = 1
_phase = 0

for theta in theta_range:
    ft = amplitude * np.sin(2 * PI * frequency * time_range + theta) * np.exp(-(time_range**2))
    sinewave =  _amplitude * np.sin(2 * PI * frequency * time_range + _phase) * np.exp(-(time_range**2))
    sinewaves.append(np.dot(ft , sinewave))

plt.plot(theta_range, sinewaves)
plt.xlabel("Phase")
plt.ylabel("Dot product of f(t) and sines")
plt.title("Dot product of f(t) and sines")
plt.show()
```
