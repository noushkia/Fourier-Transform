# Dot/Complex-Dot Product



# Part 2A 
[1]The statistical interpretation of dot product is the sum of the covariance of the vectors, this is a measure of how strongly two vectors share their pattern of values. <br>

If the vectors are identical every value is squared, if they differ the result will be smaller than the square of the larger value. This means that the resulting value is useful for measuring how similar two vectors are.<br>

If the two vectors are orthogonal they share no common variation and the sum of the products cancel out resulting in zero.
